```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from my_tools import my_LinearRegression


from sklearn.datasets import make_regression
from sklearn.datasets import fetch_openml
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

### Test on my own simulated Data

```python
N_SAMPLES = 100
N_FEATS = 1
```

```python
### Using a ramdom points sampled uniformly from a line and adding gaussian noise to generate simulated data


X = np.random.rand(N_SAMPLES)*100 -50


#params of line
m = 3
b = 1


y = X*m + b


#params of gaussian noise
center = 0
std = 10

noise = np.random.normal(center,std,N_SAMPLES)

#mix line and noise
y = y + noise


#plot
fig2, ax2 = plt.subplots()
ax2.scatter(X,y)
plt.show()
```

```python
### using two gaussians, one bigg in front of the other
#params of gaussian sampling
center1 = 0
std1 = 100
X = np.random.normal(center1,std1,N_SAMPLES)


#params of line
m = 3
b = 1


y = X*m + b


#params of gaussian noise
center2 = 0
std2 = 10

noise = np.random.normal(center2,std2,N_SAMPLES)

#mix line and noise
y = y + noise


#plot
fig3, ax3 = plt.subplots()
ax3.scatter(X,y)
plt.show()
```

```python
reg = my_LinearRegression()

reg.Fit(X,y)

min_X = np.min(X)
max_X = np.max(X)

    

print(m,b)
print(reg.slopes_,reg.coeff_)
plt.scatter(X,y,color = "black")
plt.plot(np.linspace(min_X,max_X,N_SAMPLES), reg.Predict(np.linspace(min_X,max_X,N_SAMPLES)), color = "red") 
plt.show()
```

### Test on Scikit-learn data samples

```python
### Using sklearn to generate simulated data 

# params
noise = 10

#Generate
X, y = make_regression(N_SAMPLES, N_FEATS, noise=noise)


#plot
fig, ax = plt.subplots()
ax.scatter(X,y)
plt.show()
```

```python

regsk = my_LinearRegression()

regsk.Fit(X,y)

print(regsk.slopes_,regsk.coeff_)

min_X = np.min(X)
max_X = np.max(X)

    

plt.scatter(X,y,color = "black")
plt.plot(np.linspace(min_X,max_X,N_SAMPLES), regsk.Predict(np.linspace(min_X,max_X,N_SAMPLES)), color = "red") 
plt.show()
```

```python
### California Housing 


X, y = fetch_openml(
    name="california_housing",
    version=1,
    as_frame=False,
    return_X_y=True
)



my_reg = my_LinearRegression()
my_reg.Fit(X,y)
my_pred = my_reg.Predict(X)

print('my_score',r2_score(y,my_pred))

sk_reg = LinearRegression()
sk_reg.fit(X,y)
sk_pred = sk_reg.Predict(X)

print('sk_score',r2_score(y,sk_pred))
```
